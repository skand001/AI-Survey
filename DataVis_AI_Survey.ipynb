### Function to plot Pie Charts of the Categorical Values

```python
# Define a function to draw pie charts for a list of column names
def draw_pie_charts(data, col_names):
    for col_name in col_names:
        cat = data[col_name].value_counts()
        labels = cat.index.tolist()
        counts = cat.tolist()

        explode = (0.03,) * len(labels)
        
        plt.pie(counts, 
                labels=labels, 
                explode=explode, 
                autopct='%1.1f%%', 
                shadow=True,
                wedgeprops={'edgecolor': 'grey'}, 
                startangle=140)
        plt.title(col_name, pad=10)  
        plt.show()

col_names = ['AI Decision Importance',
    'AI Bias Concern',
    'AI Trust Level',
    'Black Box Comfort',
    'AI Past Interactions',
    'Age Group',
    'Education Level',
    'Employment Status'
]

draw_pie_charts(data, col_names)
```

### Cross Tabulation 

```python
rows = []
for idx, row in data.iterrows():
    industries = row['Expected AI Breakthroughs'].split(', ')
    for industry in industries:
        rows.append([industry, row['AI Knowledge Rating']])  

# Convert the list of rows into a DataFrame
split_df = pd.DataFrame(rows, columns=['Industry', 'AI Knowledge Rating'])

# Now you can perform a cross-tabulation on this new DataFrame
industry_crosstab = pd.crosstab(split_df['Industry'], split_df['AI Knowledge Rating'])
```

```python
rows = []
for idx, row in data.iterrows():
    explainabilities = row['AI Explainability Options'].split(', ')
    for explainability in explainabilities:
        rows.append([explainability, row['AI Knowledge Rating']])  

# Convert the list of rows into a DataFrame
split_df = pd.DataFrame(rows, columns=['Explainability', 'AI Knowledge Rating'])
explainability_AIknowledge_crosstab = pd.crosstab(split_df['Explainability'], split_df['AI Knowledge Rating'])
```

### Cross Tabulation with Heatmap Matrix Plots

```python
ai_decision_knowledge_crosstab = pd.crosstab(
    data["AI Decision Importance"], 
    data['AI Knowledge Rating'])

trust_age_crosstab = pd.crosstab(
    data["AI Trust Level"], 
    data['Age Group'])

decision_trust_crosstab = pd.crosstab(
    data["AI Decision Importance"],
    data['AI Trust Level']
)

breakthrough_selfKnowledge_crosstab = pd.crosstab(
    data["Expected AI Breakthroughs"],
    data['AI Knowledge Rating']
)

pastIntercation_blackbox_crosstab = pd.crosstab(
    data["AI Past Interactions"],
    data['Black Box Comfort']
)

biasConcern_Trust_crosstab = pd.crosstab(
    data["AI Bias Concern"],
    data['AI Trust Level']
)

Education_AI_Trustlevel_crosstab = pd.crosstab(
    data["Education Level"],
    data["AI Trust Level"]
)
industry_expectaction_AI_knowledge_crosstab = pd.crosstab(
    data["Expected AI Breakthroughs"],
    data["AI Knowledge Rating"]
)

cross_tabs = {
    'AI decision Importance vs AI Knowledge': ai_decision_knowledge_crosstab,
    'Trust in AI vs Age Group': trust_age_crosstab,
    'Trust in AI in AI outcome now vs Trust if AI systems are fully understood ': decision_trust_crosstab,
    'AI Breakthrough vs AI Knowledge': breakthrough_selfKnowledge_crosstab,
    'Past Interaction vs Black box comfort': pastIntercation_blackbox_crosstab,
    'Bias Concern vs Trust in AI': biasConcern_Trust_crosstab,
    'Education vs Trust in AI': Education_AI_Trustlevel_crosstab,
    'Expected AI Breakthroughs vs AI Knowledge': industry_expectaction_AI_knowledge_crosstab,
    'AI_Explainability vs AI Knowledge': explainability_AIknowledge_crosstab,
}

# Asked for help from ChatGPT v3.5 to create this function with Seaborn
for title, cross_tab in cross_tabs.items():
    plt.figure(figsize=(10, 8))  
    heatmap_ax = sns.heatmap(cross_tab, 
                             annot=True, 
                             fmt='d', 
                             cmap='YlGnBu', 
                             cbar=True) 
    heatmap_ax.set_title(title.replace('_', ' '), fontweight='bold')  # Make title more readable

    x_labels = heatmap_ax.get_xticklabels()
    for label in x_labels:
        if len(label.get_text()) > 1:
            label.set_rotation(45)  # rotate the label if only if label length > 1 character
    
    y_labels = heatmap_ax.get_yticklabels()
    heatmap_ax.set_yticklabels(y_labels, rotation=0)  
    plt.tight_layout()
    plt.show()
```

### Mapping Ordinal Features from Categorical to Numerical Datatype

```python
# Define your mappings
importance_mapping = {
    'Not at all important': 1,
    'Slightly important': 2,
    'Moderately important': 3,
    'Very important': 4,
    'Extremely important': 5
}
bias_concern_mapping = {'Not at all concerned': 1,
                        'Slightly concerned': 2,
                        'Moderately concerned': 3,
                        'Very concerned': 4,
                        'Extremely concerned': 5
                       }

trust_mapping = {'Definitely not': 1,
                 'Probably not': 2,
                 'Neutral': 3,
                 'Probably yes': 4,
                 'Definitely yes': 5
                }

black_box_feeling_mapping = {'Extremely uncomfortable': 1,
                             'Somewhat uncomfortable': 2,
                             'Neutral': 3,
                             'Somewhat comfortable': 4,
                             'Extremely comfortable': 5
                            }


# Convert ordinal categories to numbers
data['Understanding AI Importance'] = data['AI Decision Importance'].map(importance_mapping)
data['Bias Concern'] = data['AI Bias Concern'].map(bias_concern_mapping)
data['Trust in AI if fully understood'] = data['AI Trust Level'].map(trust_mapping)
data['Black Box Feeling'] = data['Black Box Comfort'].map(black_box_feeling_mapping)

data
```
